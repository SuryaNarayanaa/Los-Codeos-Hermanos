```python
class DisjointSet:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [0] * n

    def find(self, v):
        if self.parent[v] != v:
            self.parent[v] = self.find(self.parent[v])  # Path compression
        return self.parent[v]

    def union(self, u, v):
        root_u = self.find(u)
        root_v = self.find(v)

        if root_u != root_v:
            if self.rank[root_u] > self.rank[root_v]:
                self.parent[root_v] = root_u
            elif self.rank[root_u] < self.rank[root_v]:
                self.parent[root_u] = root_v
            else:
                self.parent[root_v] = root_u
                self.rank[root_u] += 1

def kruskal(n, edges):
    edges.sort(key=lambda x: x[2])  # Sort edges by weight
    ds = DisjointSet(n)
    mst = []
    
    for u, v, weight in edges:
        if ds.find(u) != ds.find(v):  # Check if adding the edge creates a cycle
            ds.union(u, v)
            mst.append((u, v, weight))

    return mst

# Given edges
edges = [
    (0, 1, 10), (0, 2, 6), (0, 3, 5),
    (1, 3, 15), (2, 3, 4)
]

# Number of vertices (assuming max vertex index + 1)
num_vertices = 4

# Finding MST
mst = kruskal(num_vertices, edges)
print("Minimum Spanning Tree:", mst)
```

```python

import heapq

def prim(n, adj):
    """
    Implements Prim's algorithm to find the Minimum Spanning Tree (MST).

    Parameters:
        n   : Number of vertices (1-indexed)
        adj : Adjacency list where adj[u] contains (v, weight) pairs

    Returns:
        total_weight : Total weight of the MST
        mst_edges    : List of edges in the MST
    """

    min_heap = [(0, 1, -1)]  # (weight, current_node, parent_node)
    visited = [False] * (n + 1)  # To track visited nodes
    mst_edges = []
    total_weight = 0

    while min_heap and len(mst_edges) < n - 1:
        weight, u, parent = heapq.heappop(min_heap)

        if visited[u]:
            continue

        visited[u] = True
        total_weight += weight

        if parent != -1:
            mst_edges.append((parent, u, weight))

        for v, w in adj[u]:  # Explore neighbors
            if not visited[v]:
                heapq.heappush(min_heap, (w, v, u))

    return total_weight, mst_edges
```

```python
import heapq

def prims(n,adj):
    visited = [False]*(n+1)
    mst = []
    tw = 0
    min_heap = [(0,1,-1)]  # w,u,v

    while min_heap and len(mst) < n:

        w,u,v = heapq.heappop(min_heap)

        if   visited[u]:
            continue

        visited[u] = True
        tw+=w

        if v!= -1:
            mst.append((v,u,w))

        for v,w in adj[u]:
            if not visited[v]:
                heapq.heappush(min_heap ,(w,v,u))

    return tw, mst



# Example Usage
n, m = 6, 7  # Number of nodes and edges
edges = [
    (1, 6, 10), (1, 3, 6), (1, 4, 5),
    (2, 4, 15), (3, 4, 4)
]
# Construct adjacency list
adj = {i: [] for i in range(1, n + 1)}
for u, v, w in edges:
    adj[u].append((v, w))
    adj[v].append((u, w))  # Undirected graph

total_weight, mst_edges = prims(n, adj)

print("Minimum Spanning Tree edges:")
for u, v, w in mst_edges:
    print(f"{u} -- {v} : {w}")
print("Total Weight of MST:", total_weight)
```
