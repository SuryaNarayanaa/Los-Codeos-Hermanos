```python
# Create a palindrome linked list
palindrome_list = LinkedList()

# Adding values to create a palindrome: 1->2->3->2->1
palindrome_list.append(1)
palindrome_list.append(2)
palindrome_list.append(3)
palindrome_list.append(2)
palindrome_list.append(1)

# Print the palindrome list
print("Palindrome linked list:")
palindrome_list.print_list()

# To verify it's actually a palindrome, you can add a method to check
def is_palindrome(head):
    # Convert linked list to array
    values = []
    current = head
    while current:
        values.append(current.data)
        current = current.next
    
    # Check if array is palindrome
    return values == values[::-1]

print(f"Is palindrome: {is_palindrome(palindrome_list.head)}")
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.prev = None
        self.next = None
        self.child = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, parent, val):
        """Appends a new node at the end of the given parent level."""
        new_node = Node(val)
        if not parent:
            self.head = new_node
            return new_node

        current = parent
        while current.next:
            current = current.next
        current.next = new_node
        new_node.prev = current
        return new_node

    def add_child(self, parent, val):
        """Adds a child node to a parent node."""
        if not parent:
            return None
        parent.child = Node(val)
        return parent.child

    def print_list(self, node, level=0):
        """Prints the multi-level linked list structure."""
        while node:
            print(" " * (level * 4) + f"{node.val}", end=" -> ")
            if node.child:
                print("\n" + " " * (level * 4) + "|")
                self.print_list(node.child, level + 1)
            node = node.next
        print("None")

    def flatten(self, head):
        
        stack = []
        curr = head
        while(curr):
            if curr.child is not None:
                stack.append(curr)
                curr = curr.child
            else:
                if curr.next is None and stack :
                    new_curr= stack.pop()
                    t = new_curr.next
                    new_curr.next.prev = curr
                    new_curr.next = new_curr.child
                    
                    curr.next = t
                    new_curr.next.prev = new_curr
                    
                    curr = curr.next

                else:
                    curr = curr.next

        return head
# Creating the multi-level doubly linked list
dll = DoublyLinkedList()

# Level 1
n1 = dll.append(None, 1)
n2 = dll.append(n1, 2)
n3 = dll.append(n2, 3)
n4 = dll.append(n3, 4)
n5 = dll.append(n4, 5)
n6 = dll.append(n5, 6)

# Level 2
n7 = dll.add_child(n3, 7)
n8 = dll.append(n7, 8)
n9 = dll.append(n8, 9)
n10 = dll.append(n9, 10)

# Level 3
n11 = dll.add_child(n8, 11)
n12 = dll.append(n11, 12)

# Print the list before flattening
print("Multi-Level Doubly Linked List:")
dll.print_list(dll.head)

# Flatten the list
dll.flatten(dll.head)
print("\nFlattened Doubly Linked List:")
dll.print_list(dll.head)
```
