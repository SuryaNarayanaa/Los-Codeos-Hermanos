```python
board = [["5","3",".",".","7",".",".",".","."],["6",".",".","1","9","5",".",".","."],[".","9","8",".",".",".",".","6","."],["8",".",".",".","6",".",".",".","3"],["4",".",".","8",".","3",".",".","1"],["7",".",".",".","2",".",".",".","6"],[".","6",".",".",".",".","2","8","."],[".",".",".","4","1","9",".",".","5"],[".",".",".",".","8",".",".","7","9"]]
```

```python
def show_board(b):
    r,c = len(b) , len(b[0])

    for i in range(r): 
        if (i)%3 == 0 and i!=0 :
            print("-----------------------------")
        for j in range(c):
            if j != 8:
                print(str(b[i][j])+" ", end="")
            else:
                print(str(b[i][j]), end="")

            if (j+1)%3 == 0 and j!=0 :
                print("|", end="")
        print("\n")
show_board(board)
```

```python
def find_pos(b):
    for i in range(len(b)):
        for j in range(len(b[0])):
            if b[i][j] == ".":
                return (i,j)
    return None
```

```python
def check(b,pos ,num):
    r , c = len(b), len(b[0])
    x,y = pos
    for i in range(r):
        if x!=i and b[i][y] == num:
            return False
    for i in range(c):
        if y!=i and b[x][i] == num:
            return False
    
    x_b , y_b = x//3 , y//3

    for i in range(x_b*3 , x_b*3 +3):
        for j in range(y_b*3 , y_b*3 +3):
            if b[i][j] == num and i!= x  and j!=y:
                return False
    return True

check( [["5","3","4","6","7","8","9","1","2"],["6","7","2","1","9","5","3","4","8"],["1","9","8","3","4","2","5","6","7"],["8","5","9","7","6","1","4","2","3"],["4","2","6","8","5","3","7","9","1"],["7","1","3","9","2","4","8","5","6"],["9","6","1","5","3","7","2","8","4"],["2","8","7","4","1","9","6","3","5"],["3","4","5","2","8","6","1","7","9"]], (8,8), 9)
```

```python
def solve(b):
    if not find_pos(b) :
        return True
    else:
        pos = find_pos(b)
        for i in range(1,10):
            if check(b,pos, str(i)):
                b[pos[0]][pos[1]] = str(i)
                if solve(b):
                    return True
                else:
                    b[pos[0]][pos[1]] = "."
    return False
solve(board)
show_board(board)
```
